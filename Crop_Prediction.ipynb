```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("Crop_recommendationnn.csv")
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df["label"].value_counts()
list(df['label'].unique())
```

```python
df.groupby("label").mean()
```

```python
# Separating the data and labels
x = df.drop(columns="label", axis=1)
y = df["label"]
x.columns.values
```

## Data preprocessing

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x)
```

```python
scaled_x = scaler.transform(x)
```

```python
scaled_x
```

```python
encoder = LabelEncoder()
df["label"] = encoder.fit_transform(df["label"])
print(encoder.classes_)
df.head()
```

```python
x = scaled_x
y = df["label"]
```

## Train test split

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size = .2, stratify = y, random_state = 1)
```

```python
print(x.shape, xtrain.shape, xtest.shape)
```

## Model Training

```python
svm_model = SVC()
nb_model = GaussianNB()
rf_model = RandomForestClassifier(random_state=1)
svm_model.fit(xtrain, ytrain)
nb_model.fit(xtrain,ytrain)
rf_model.fit(xtrain,ytrain)
```

```python
svm_preds = svm_model.predict(xtrain)
nb_preds = nb_model.predict(xtrain)
rf_preds = rf_model.predict(xtrain)
rf_preds
```

```python
from statistics import mode
final_preds = [mode([i,j,k]) for i,j,k in zip(svm_preds, nb_preds, rf_preds)]
final_preds
```

```python
# Accuracy score on training data
training_data_accuracy = accuracy_score(svm_preds,ytrain)
training_data_accuracy
```

```python
# Accuracy score on training data
training_data_accuracy = accuracy_score(nb_preds,ytrain)
training_data_accuracy
```

```python
# Accuracy score on training data
training_data_accuracy = accuracy_score(rf_preds,ytrain)
training_data_accuracy
```

```python
# Accuracy score on training data
training_data_accuracy = accuracy_score(final_preds,ytrain)
training_data_accuracy
```

```python
# test data prediction
svm_preds = svm_model.predict(xtest)
nb_preds = nb_model.predict(xtest)
rf_preds = rf_model.predict(xtest)
```

```python
final_preds = [mode([i,j,k]) for i,j,k in zip(svm_preds, nb_preds, rf_preds)]
final_preds
```

```python
training_data_accuracy = accuracy_score(svm_preds,ytest)
training_data_accuracy
```
